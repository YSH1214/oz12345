Git의 기본 설정

Git을 이용한 프로젝트
1. 프로젝트 작업을 위한 폴더 생성
2. VScode를 이용해 해당 폴더 열기
3. 터미널 열기
4. 전역 설정하기

    [입력]
    - git config --global user.name 닉네임
    - git config --global user.email github에 등록한 이메일
    - git config --global init.defaultBranch main

    [삭제]
    - git config --global --unset user.name
    - git config --global --unset user.email

    [내용확인]
    - git config --global user.name
    - git config --global user.email
    - git config --list 

5. Git이 관리하는 폴더로 설정하기

    - git init

        - git 관리하는 폴더로 지정되면서 현재 작업하고 있는 공간을 main으로 지칭
        - .git이라는 숨겨진 폴더가 생김, git에서 발생하는 모든 정보는 해당 폴더안에서 관리됨

6. Untracked와 Staged

    - Untracked : 새로 생성된 파일의 상태(빨간색)
    - Staged : commit이 가능한 위치(초록색)

7. ignore 파일

    - ignore 파일 생성해 버전(commit)으로 만들고 싶지 않은 파일 관리하기
        - .gitignore 파일 생성
        - gitignore 파일 안에 버전(commit)으로 만들고 싶지 않은 파일명 기재

8. git add

    [Add]
    - git add 파일명 : 원하는 하나의 파일을 Staged에 올리는 명령어
    - git add . : 원하는 모든 파일을 Staged에 올리는 명령어

    [reset]
    - git reset : Staged에 올라가 있는 모든 파일을 Untracked 상태로 되돌리는 명령어

9. git의 Working Directory의 상태 확인 명령어

    - git status : 현재 git이 관리하는 공간에서 발생한 일들을 한눈에 볼 수 있는 명령어

10. git commit

    - git commit -m "커밋 메시지" : commit 명령어를 통해<br>
    그 순간을 사진으로 찍듯 Version을 만드는 명령어
    - git commit : -m 옵션 없이 git commit 명령어만 입력한 경우 Vim 모드로 진입<br>
    "i"를 눌러 입력모드로 전환 후 상단에 commit 메시지 입력 "ESC"를 눌러 입력 모드 해제<br>
    "Shift + ;"를 눌러 명령어 입력모드로 전환 후 "wq"입력해 저장하고 나가기

11. git log

    - commit 이력을 확인하는 명령어
    - git log --oneline
